### Get properties from Vísir

```python
import requests
from random import sample

empty_arr = []

for i in range(1, 6):
    r = requests.get('http://fasteignir.visir.is/api/search?onpage=1000&page='+str(i))
    if r is not None:
        empty_arr.append(r.json())


# Initialize an empty array
arr = []
for inner_array in empty_arr:
    for dictionary in inner_array:
        arr.append(dictionary)
```

### Working with the data

```python
from datetime import datetime
def check_if_property_has_been_sold_before(address, zip_number, size):
    previosly_sold_for = []
    if (kaupskra['heimilisfang'].apply(str.lower).eq(address.lower())).any():
        this_address = kaupskra[kaupskra['heimilisfang'].apply(str.lower) == address.lower()]
        
        this_address_and_zip = this_address[this_address["postnr"]==int(zip_number)]
        if this_address_and_zip.shape[0] > 0: # check if correct property based on the zip for property
            with_this_size = this_address_and_zip[this_address_and_zip["einflm"]==size]
            if with_this_size.shape[0] > 0: # check if same fm
                # df_to_use = with_this_size.copy()
                # df_to_use['utgdag'] = pd.to_datetime(df_to_use['utgdag'])
                # df_to_use = df_to_use.sort_values(by='utgdag', inplace = True) 
                
                previosly_sold_for = with_this_size["kaupverd"].div(1000).round(2).to_numpy()
                utgafu_dagur = with_this_size["utgdag"].to_numpy()
                for p in range(0, len(previosly_sold_for)):
                    sold_date = datetime.strptime(utgafu_dagur[p], '%Y-%m-%d %H:%M:%S.%f').strftime('%d %B %Y')
                    print("Previous price: "+ str(previosly_sold_for[p]) + " milljonir ("+ sold_date + ")")
        
def print_out_details(items):
    for item in items:
        print(item)
```

```python
show_top_n = 25
for item in arr[0:show_top_n]:
    # values from columns inside the printable string (if applies)
    address = item["street_name"] + " " +  item["street_number"]
    size = float(item["size"].replace(",", "."))
    zip = item["zip"]["zip"]
    price_in_millions_isk = int(item["price"])/1000000

    # Texts
    TEXT_for_sale_or_rent = "Til sölu" if item["sale_or_rent"] == '1' else "Til leigu"
    TEXT_size = "Fermeterar: " + str(size)
    TEXT_price = f"Verð: {str(price_in_millions_isk)} milljónir" if float(price_in_millions_isk) != 0.0 else f"Verð: Tilboð"
    print_out_details(items=[address, TEXT_for_sale_or_rent, TEXT_size,TEXT_price])
    
    # check kaupskra.csv for if this property has been sold previously
    check_if_property_has_been_sold_before(address=address, zip_number=zip, size=size)
    print("\n")
```
