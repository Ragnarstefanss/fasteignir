```python
import requests
from random import sample
empty_arr = []

for i in range(1, 6):
    r = requests.get('http://fasteignir.visir.is/api/search?onpage=1000&page='+str(i))#.json()
    if r is not None:
        empty_arr.append(r.json())


# Initialize an empty array
arr = []
for inner_array in empty_arr:
    for dictionary in inner_array:
        arr.append(dictionary)
```

```python
# Look lets look at one sample to see how the data looks like
print(arr[0].keys()) #primary keys but not the nested keys
sample(arr, 1)
```

```python
for item in arr:
    heimilisfang = item["street_name"] + " " +  item["street_number"]
    sala_eda_leiga = "Til sölu" if item["sale_or_rent"] == '1' else "Til leigu"
    fermetrar = "Fermeterar: " + item["size"]
    
    verd_i_milljonum = str(int(item["price"])/1000000)
    prentad_verd = f"Verð: {verd_i_milljonum} milljónir" if float(verd_i_milljonum) != 0.0 else f"Verð: Tilboð"
    
    print(heimilisfang + "\n" + sala_eda_leiga + "\n" + fermetrar + "\n" + prentad_verd+"\n\n")
```
